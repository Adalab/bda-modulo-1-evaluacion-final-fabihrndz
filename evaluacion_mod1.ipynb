```python
class TiendaOnline:    #Clase
    
    def __init__(self):  #Constructor
            #Atributos
        self.inventario = []       #Lista para el inventario vacia      
        self.ventas_totales = 0.0    #Para almacenar las ventas totales(float)

                 #Metodos
    #Agregar producto, si existe se suma, si no existe se agrega
    def agregar_producto(self, nombre, precio, cantidad):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                producto["cantidad"] += cantidad
                return   #return para que no se ejecute lo de abajo (si se ejecuta el if)
            #Se ejecuta si no entra en el if    
        self.inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})
    #Ver el inventario
    def ver_inventario(self):
        for producto in self.inventario:                 
            print(f" Nombre: {producto["nombre"]}, Precio: {producto["precio"]}, Cantidad: {producto["cantidad"]}")
    #Buscar por nombre del producto
    def buscar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"] == nombre:  
                print(f" Nombre: {producto["nombre"]}, Precio: {producto["precio"]}, Cantidad: {producto["cantidad"]}")
                return            
        print("El producto no existe en el inventario")
    #Actualizar las cantidades de producto que tengamos en el stock
    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                producto["cantidad"] += cantidad
                return  #Si entra en el if, termina la ejecucion
        print("El producto no esta en el inventario")  #Si no entra en el if
    #Eliminar producto por nombre
    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                self.inventario.remove(producto)
                return #Finaliza aqui
        print("No se puede eliminar, no esta en el inventario") #No entra al if, ejecuta el print
        #Calcular el total del inventario
    def calcular_valor_inventario(self):
        total = 0     #Para que almacene cada iteracion
        for producto in self.inventario:
            total += (producto["cantidad"] * producto["precio"])
        print(f"El valor total del stock es {total}€")
```
