```python
class TiendaOnline:    #Clase
    
    def __init__(self):  #Constructor
            #Atributos
        self.inventario = []       #Lista para el inventario vacia      
        self.ventas_totales = 0.0    #Para almacenar las ventas totales(float)
        self.clientes = {}

                 #Metodos
    #Agregar producto, si existe se suma, si no existe se agrega
    def agregar_producto(self, nombre, precio, cantidad):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                producto["cantidad"] += cantidad
                return   #return para que no se ejecute lo de abajo (si se ejecuta el if)
            #Se ejecuta si no entra en el if    
        self.inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})
    #Ver el inventario
    def ver_inventario(self):
        for producto in self.inventario:                 
            print(f" Nombre: {producto["nombre"]}, Precio: {producto["precio"]}, Cantidad: {producto["cantidad"]}")
    #Buscar por nombre del producto
    def buscar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"] == nombre:  
                print(f" Nombre: {producto["nombre"]}, Precio: {producto["precio"]}, Cantidad: {producto["cantidad"]}")
                return            
        print("El producto no existe en el inventario")
    #Actualizar las cantidades de producto que tengamos en el stock
    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                producto["cantidad"] += cantidad
                return  #Si entra en el if, termina la ejecucion
        print("El producto no esta en el inventario")  #Si no entra en el if
    #Eliminar producto por nombre
    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre"] == nombre:
                self.inventario.remove(producto)
                return #Finaliza aqui
        print("No se puede eliminar, no esta en el inventario") #No entra al if, ejecuta el print
    #Calcular el total del inventario
    def calcular_valor_inventario(self):
        total = 0     #Para que almacene cada iteracion
        for producto in self.inventario:
            total += (producto["cantidad"] * producto["precio"])
        print(f"El valor total del stock es {total}€")

                        #BONUS
        
    #Agregar clientes al diccionario
    def agregar_cliente(self, nombre, email):
        self.clientes[nombre] = {"email": email}  #Agregamos cliente al dicionario con otro diccionario para almacenar el email
        print(f"Cliente añadido con exito")

    # Ver una lista de los clientes
    def ver_clientes(self):
        for k,v in self.clientes.items(): #Itera en el diccionario
            print(f"Nombre: {k}, Correo: {v["email"]}")  #Accede a su clave (nombre) y dentro del diccionario accede a la clave email 

    # Permitir que el cliente compre varios articulos con un bucle, sumar al carrito, realizar compra, restar al inventario
    def realizar_compra(self):
        carrito = {} #El carro de la comprar
        print(self.ver_inventario()) #Mostrar el inventario
        while True:           #Bucle
            compra = input("Ingresa el nombre del producto que deseas o pulsa (S) para salir").lower() 
            if compra == "s":        #Permite salir del bucle
                print("¡¡Gracias por la visita, hasta luego!!")
                break
            for producto in self.inventario:  #Recorrer los diccionarios dentro de la lista
                if compra == producto["nombre"]:   #Verificar que el producto esta en el inventario
                    if producto["cantidad"] > 0:   #Verificar que tiene stock
                        if compra in carrito:
                            carrito[compra]["cantidad"] += 1          #Si ya habia añadido alguno al carrito se suma la cantidad
                        else:                                         #Si no habia añadido al carrito, se añade
                            carrito[compra] = {"precio": producto["precio"], "cantidad": 1}
                    else:                                #Else para el if que verifica si hay stock disponible
                        print("Ingresa un articulo con stock")
                    break
            else:              #for-else por si no entra en el if
                print("El producto no esta en el stock")
        #Restar del inventario
        for producto in self.inventario:      #Iterar por la lista
            if producto["nombre"] in carrito:   #Si el producto esta en el carrito
                producto["cantidad"] -= carrito[producto["nombre"]]["cantidad"]    #Resta la cantidad al inventario
        #Mostrar carrito
        total = 0
        print("Compra realizada")
        for k,v in carrito.items():
            subtotal = v["precio"] * v["cantidad"]
            total += subtotal
            print(f"{producto["nombre"]}, Cantidad: {v["cantidad"]}, Subtotal: {subtotal}")
            print(f"Total de la compra: {total}")                  
```
